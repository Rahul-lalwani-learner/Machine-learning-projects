```python
IMGPATH="dataset"
CATEGORIES=['abi','adi','adu','akshada','arin','harsh','piyush','prince','rahul','rohit','sahil','samriddhi']
# abi=[]
IMGSIZE=200
for category in CATEGORIES: 
    path=os.path.join(IMGPATH,category)
    for img in os.listdir(path): 
        img_array=cv.imread(os.path.join(path,img),cv.IMREAD_GRAYSCALE)
        cropped_img=img_array[2400:3725,725:2200]
        new_array=cv.resize(cropped_img,(IMGSIZE,IMGSIZE))
#         plt.imshow(img_array,cmap='gray')
#         abi.append(cropped_img)
        plt.imshow(new_array,cmap='gray')
        plt.show()
        break
    break
# len(abi)
new_array.shape
```

```python
wavelength = 5
orientation = 0
phase_offset = 1*np.pi
sigma = 3

for category in CATEGORIES: 
    path=os.path.join(IMGPATH,category)
    for img in os.listdir(path): 
        img_array=cv.imread(os.path.join(path,img),cv.IMREAD_GRAYSCALE)
        cropped_img=img_array[2400:3725,725:2200]
        new_array=cv.resize(cropped_img,(IMGSIZE,IMGSIZE))
        g_kernel = cv.getGaborKernel((wavelength, wavelength), sigma, orientation, wavelength, 0.5, 0)
        fimage = cv.filter2D(new_array, cv.CV_8UC3, g_kernel)
        plt.imshow(fimage,cmap='gray')
        plt.show()
        break
    break
```

```python
data_meanth=[]
for category in CATEGORIES: 
    path=os.path.join(IMGPATH,category)
    label_index=CATEGORIES.index(category)
    for img in os.listdir(path): 
        img_array=cv.imread(os.path.join(path,img),cv.IMREAD_GRAYSCALE)
        cropped_img=img_array[2400:3725,725:2200]
        new_array=cv.resize(cropped_img,(IMGSIZE,IMGSIZE))
        g_kernel = cv.getGaborKernel((wavelength, wavelength), sigma, orientation, wavelength, 0.5, 0)
        fimage = cv.filter2D(new_array, cv.CV_8UC3, g_kernel)
        mblur=cv.medianBlur(fimage,5)
        meanth=cv.adaptiveThreshold(mblur,255,cv.ADAPTIVE_THRESH_MEAN_C,cv.THRESH_BINARY,11,2)
        data_meanth.append([meanth,label_index])
        
```

```python
data_gaussianth=[]
for category in CATEGORIES: 
    path=os.path.join(IMGPATH,category)
    label_index=CATEGORIES.index(category)
    for img in os.listdir(path): 
        img_array=cv.imread(os.path.join(path,img),cv.IMREAD_GRAYSCALE)
        cropped_img=img_array[2400:3725,725:2200]
        new_array=cv.resize(cropped_img,(IMGSIZE,IMGSIZE))
        g_kernel = cv.getGaborKernel((wavelength, wavelength), sigma, orientation, wavelength, 0.5, 0)
        fimage = cv.filter2D(new_array, cv.CV_8UC3, g_kernel)
        mblur=cv.medianBlur(fimage,5)
        gaussianth=cv.adaptiveThreshold(mblur,255,cv.ADAPTIVE_THRESH_GAUSSIAN_C,cv.THRESH_BINARY,11,2)
        data_gaussianth.append([gaussianth,label_index])
        
        
```
